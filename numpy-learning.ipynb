```python
import numpy as np

a = np.arange(15).reshape(3,5)
print(f"a: {str(a.tolist()):>69}\n"
      f"Shape of a: {str(a.shape):>10}\n"
      f"Dims of a: {str(a.ndim):>10}\n"
      f"Data type of a: {str(a.dtype.name):>9}\n"
      f"Item size a: {str(a.itemsize):>10}\n"
      f"Size of a: {str(a.size):>10}\n"
      f"type of a: {str(type(a)):>10}\n"
     )
```
